Importing Gold Sample

```
from astropy.io import fits
from astropy.table import QTable, join
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib as mpl
gold_sample=QTable.read("https://users.flatironinstitute.org/~apricewhelan/data/dr16-binaries/gold_sample.fits")
```

Creating Scatter Plot

```
title_font= {'family': 'serif',
        'color':  'black',
        'weight': 'bold',
        'size': 20,
        }
label_font={'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 15,
        }
min=gold_sample["MAP_K"].min()/u.km*u.s
max=gold_sample["MAP_K"].max()/u.km*u.s
fig,ax=plt.subplots(1,1, figsize=(10,10),facecolor="silver")
ax.scatter(gold_sample["MAP_P"],gold_sample["MAP_e"],marker="*",
           c=gold_sample['MAP_K'],cmap="inferno_r",
           norm=colors.Normalize(vmin=min,vmax=max))
ax.set_xscale("log")
fig.colorbar(mpl.cm.ScalarMappable(colors.Normalize(vmin=min,vmax=max),
             cmap="inferno_r"),
             label="Radial Velocity [km/s]")
ax.set_title("Orbital Period Vs Eccentricity of the \"Gold Sample\"",fontdict=title_font)
ax.set_ylabel("Eccentricity",fontdict=label_font)
ax.set_xlabel("Period [d]",fontdict=label_font)
ax.set_facecolor("silver")
```
